```python
#Reconstructing strings from k-mers

def composition(seq, k):
    kmers = []
    for i in range(0, len(seq) - k +1):
        window = seq[i:i+k]
        kmers.append(window)
    return kmers

with open('dataset_30153_3.txt') as f:
    lines = [line.rstrip('\n') for line in f]
    k= int(lines[0])
    seq = lines[1]

print(" ".join(composition(seq, k)))
        
```

```python
#String Spelled by a Genome Path Problem. Reconstruct a string from its genome path.

def PathToGenome(path):
    seq = ""
    seq += path[0][:-1]
    for kmer in path:
        seq += kmer[-1]
    return seq

path = "ACCGA CCGAA CGAAG GAAGC AAGCT".split(" ")

with open('dataset_30182_3.txt') as f:
    lines = [line.rstrip('\n') for line in f]
    path = lines[0].split(" ")
 
PathToGenome(path)
        
        
        
```

```python
import numpy as np

def suffix(kmer):
    return kmer[1:]

def prefix(kmer):
    return kmer[:-1]

def adj_matrix(kmers):
    n = len(kmers)
    matrix = np.zeros(shape=(n, n))
    for i in range(0,n):
        suf = suffix(kmers[i])
        for j in range(0,n):
            pre = prefix(kmers[j])
            if suf == pre:
                matrix[i, j] = 1

    return matrix
                

def adj_list(kmers):
    n = len(kmers)
    matrix = {}
    for kmer in kmers:
        matrix[kmer] = []
        for neighbor in kmers:
            if suffix(kmer) == prefix(neighbor):
                matrix[kmer].append(neighbor)
    return matrix

def output_adj_list(matrix):
    for key in matrix.keys():
        if len(matrix[key]) > 0:
            neighbors = " ".join(matrix[key])
            print(f"{key}: {neighbors}")
            
        

kmers = "ATGCG GCATG CATGC AGGCA GGCAT GGCAC".split(" ")

with open('dataset_30182_10.txt') as f:
    lines = [line.rstrip('\n') for line in f]
    kmers = lines[0].split(" ")
 

output_adj_list(adj_list(kmers))

 
```
