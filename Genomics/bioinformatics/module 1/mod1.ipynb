```python
with open('dataset_30272_13.txt') as f:
    lines = [line.rstrip('\n') for line in f]
text = lines[0]
k = int(lines[1])

#dictonary of all kmers in text and their counts.
def FrequencyTable(text, k):
    dic = {}
    for i in range(0, len(text) - k + 1):
        pattern = text[i:i+k]
        if pattern not in dic.keys():
            dic[pattern] = 1
        else:
            dic[pattern] += 1
    return dic

#maximum dictionary value
def MaxMap(dic):
    mx = max(dic.values())
    return mx

#find the most frequent kmer (or kmers).
def BetterFrequentWords(text, k):
    freq = []
    dic = FrequencyTable(text, k)
    mx =  MaxMap(dic)
    for pattern in dic.keys():
        if dic[pattern] == mx:
            freq.append(pattern)
    return freq

freq = BetterFrequentWords(text, k)
print(" ".join(freq))
```

```python
with open('dataset_30273_2.txt') as f:
    lines = [line.rstrip('\n') for line in f]
text = lines[0]

def rc(pattern):
    complement = {'A': 'T', 'C': 'G', 'G': 'C', 'T': 'A'} 
    rc = "".join(complement.get(base, base) for base in reversed(pattern))
    return rc
```

```python
with open('dataset_30273_5.txt') as f:
    lines = [line.rstrip('\n') for line in f]

pattern = lines[0]
text = lines[1]

def findall(text, pattern):
    locations = []
    reverse = rc(pattern)
    for i in range(0, len(text) -len(pattern) +1):
        if text[i:i+len(pattern)] == pattern or text[i:i+len(pattern)] == reverse:
            locations.append(i)
    return " ".join(str(num) for num in locations)

findall(text,pattern)
```

```python
def FindClumps(text, k, L, t):
    patterns = []
    for i in range(0, len(text) - L + 1):
        window = text[i:i+L]
        dic = FrequencyTable(window, k)
        for s in dic.keys():
            if dic[s] >= t:
                patterns.append(s)
    return list(set(patterns))

with open('dataset_30274_5.txt') as f:
    lines = [line.rstrip('\n') for line in f]

text = lines[0]
k, L, t = map(int,lines[1].split(" "))

print(" ".join(FindClumps(text, k, L, t)))
```
