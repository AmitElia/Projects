```python
import torch
idx = 59
row = table_600.iloc[idx]

cols_raw = row.get("token_positions")
cols_by_site = row.get("token_positions_by_site")
cols_val = row.get("token_positions_validated_by_site")
bos = int(row.get("bos_offset", 0))

q_res = row["res_embeddings"]                          # Torch tensor
Lt   = int(q_res.shape[0])

# What the scorer would actually use
cols_used = hv1._collect_site_cols_for_row(row, allow_single_mismatch=True)

print("bos_offset:", bos, "Lt:", Lt)
print("raw union (1-based):", sorted({int(j) for v in (cols_raw or {}).values() for j in v}))
print("used (0-based):", cols_used)
print("any oob? ", any(j < 0 or j >= Lt for j in cols_used))

# Self score (Torch only; NO NumPy)
S = hv1.get_per_res_similarity_matrix(q_res, row["res_embeddings"])  # should already be CUDA
S = S.to(q_res.device)                                               # safety

if cols_used:
    idx_t = torch.as_tensor(cols_used, device=S.device, dtype=torch.long)
    v = S.index_select(1, idx_t).sum(dim=1)                          # [Lq] CUDA
    k = min(6, v.numel())
    top_vals, top_idx = torch.topk(v, k=k, largest=True)
    print("self top-6:", top_vals.tolist(), "sum:", float(top_vals.sum().item()))
else:
    v = torch.zeros(q_res.shape[0], device=q_res.device)
    print("No site columns")
```
