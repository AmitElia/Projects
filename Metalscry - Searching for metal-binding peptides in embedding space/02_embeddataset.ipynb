```python
import matplotlib.pyplot as plt
import numpy as np
import metalscry_search as ms

# config
k_list       = [5, 20]
temp_list    = [0.1, 0.5]
threshold    = 0.01          # single value (for hits only)
#marker_taus  = [0.01, 0.02, 0.03, 0.04]   # score thresholds to mark
prior_q      = None
alpha        = 0.5

summary_df, curves_loc, curves_id, markers = ms.evaluate_grid(
    test_df_x, test_df_y, train_df,
    k_list=k_list, temp_list=temp_list,
    threshold=threshold,                  # for hits only
    # explicit Ï„ to mark:
    marker_taus_loc=[0.01], marker_taus_id=[0.01],
    # add quantile-based Ï„:
    marker_quantiles_loc=[0.99, 0.95],
    marker_quantiles_id=[0.99, 0.95],
    # add target-FPR markers (especially useful on ROC with log-x):
    marker_fprs_loc=[0.10, 0.01],
    marker_fprs_id=[0.10, 0.01],
    # scoring knobs:
    use_raw_votes=True, use_mass=True, alpha=0.5, prior_q=None
)
print(summary_df)

```

```python
# plot 2x2
fig, axs = plt.subplots(2, 2, figsize=(12, 12))
(ax1, ax2), (ax3, ax4) = axs

# PR as-is
ax1, h1, l1 = ms.plot_grid_pr_ax(curves_loc, ax1, "Location")
ax3, h3, l3 = ms.plot_grid_pr_ax(curves_id,  ax3, "Identity")

# ROC zoomed + partial AUC + log-x
ax2, h2, l2 = ms.plot_grid_roc_ax(
    curves_loc, ax2, "Location",
    max_fpr=1.00, log_x=True, pauc_at=1.00, equal_square=False
)
ax4, h4, l4 = ms.plot_grid_roc_ax(
    curves_id,  ax4, "Identity",
    max_fpr=1.0, log_x=True, pauc_at=1.0, equal_square=False
)

# plotting (same as you do now) + overlay markers:
for (k, T), pts in markers["location"].items():
    for (rec, prec, tau, lab) in pts["pr"]:  ax1.scatter(rec, prec, s=40)
    for (fpr, tpr, tau, lab) in pts["roc"]:  ax2.scatter(fpr, tpr, s=40)

for (k, T), pts in markers["identity"].items():
    for (rec, prec, tau, lab) in pts["pr"]:  ax3.scatter(rec, prec, s=40)
    for (fpr, tpr, tau, lab) in pts["roc"]:  ax4.scatter(fpr, tpr, s=40)

# legend below (curves only)
handles = h1 + h2 + h3 + h4
labels  = l1 + l2 + l3 + l4
dedup = {}
for h, lab in zip(handles, labels):
    dedup.setdefault(lab, h)
H, L = list(dedup.values()), list(dedup.keys())
ncols = max(2, min(10, int(np.ceil(np.sqrt(len(L))))))
fig.legend(H, L, loc="lower center", bbox_to_anchor=(0.5, 0.02),
           ncol=ncols, frameon=False, fontsize=8, handlelength=2.0, columnspacing=1.0)
fig.subplots_adjust(bottom=0.18, wspace=0.25, hspace=0.25)
plt.show()
```
