```python
import pandas as pd
import ast
from collections import Counter
import matplotlib.pyplot as plt

def _parse_metal_cell(cell):
    """Return a dict from a cell that may be a dict, stringified dict, or NaN."""
    if isinstance(cell, dict):
        return cell
    if pd.isna(cell):
        return {}
    if isinstance(cell, str):
        try:
            obj = ast.literal_eval(cell)
            return obj if isinstance(obj, dict) else {}
        except (ValueError, SyntaxError):
            return {}
    return {}

def compute_metal_counts(df: pd.DataFrame, col: str = "metal_residues") -> pd.DataFrame:
    rows_with = Counter()
    site_counts = Counter()
    residue_counts = Counter()

    for cell in df[col]:
        d = _parse_metal_cell(cell)
        for metal, groups in d.items():
            m = str(metal).upper()
            rows_with[m] += 1  # mention = appears in this row

            # groups may be list[list[str]] (multiple sites) or list[str] (single site)
            if isinstance(groups, list):
                if groups and all(isinstance(g, list) for g in groups):
                    site_counts[m] += len(groups)
                    residue_counts[m] += sum(len(g) for g in groups)
                else:
                    site_counts[m] += 1
                    residue_counts[m] += sum(isinstance(x, str) for x in groups)
            else:
                site_counts[m] += 1

    metals = sorted(set(rows_with) | set(site_counts) | set(residue_counts))
    out = pd.DataFrame({
        "metal": metals,
        "rows_with_metal": [rows_with[m] for m in metals],
        "num_sites": [site_counts[m] for m in metals],
        "num_residues": [residue_counts[m] for m in metals],
    }).sort_values("rows_with_metal", ascending=False, ignore_index=True)
    return out

# --- run on your table ---
metal_counts = compute_metal_counts(final, "metal_residues")

# Plot "mentions" = rows that include each metal at least once
plt.figure(figsize=(10,5))
plt.bar(metal_counts["metal"], metal_counts["rows_with_metal"])
plt.ylabel("Rows mentioning metal")
plt.xlabel("Metal")
plt.title("Metal mentions per row")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()

# Optional: view the counts table
metal_counts
```
